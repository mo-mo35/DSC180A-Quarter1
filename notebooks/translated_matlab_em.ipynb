```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal, expon
import math

# Find next prime number
def nextprime(n):
    def isprime(x):
        if x < 2:
            return False
        for i in range(2, int(math.sqrt(x)) + 1):
            if x % i == 0:
                return False
        return True
    
    while not isprime(n):
        n += 1
    return n

# Set the dimensions of the problem
n = int(nextprime(20))  # size of observations
d = int(nextprime(3))   # number of latent dimensions
T = int(nextprime(int(1e3)))  # number of trials

# Specify SNR and noise variance
SNR = 0.5
nseVar = d**2 / SNR

# Generate data
W = np.random.randn(n, d)  # weight matrix
z = np.random.randn(d, T)  # latents
lamb = expon.rvs(scale=1, size=n) / nseVar  # noise precisions
Y = W @ z + np.diag(np.sqrt(1.0 / lamb)) @ np.random.randn(n, T)

# Do EM (with help from Bishop)
I = 1200  # number of iterations REDUCE TO DECREASE COMPUTATION TIME
Wn = np.random.randn(n, d)  # Initialize new W
lambNew = nseVar * expon.rvs(scale=1, size=n)  # initialize noise variances
S = (Y @ Y.T) / T  # raw data sample covariance
ll = np.full(I, np.nan)

for ii in range(I):
    # reset
    Wo = Wn;
    lambOld = lambNew;

    # E-step
    invPsiOld = np.diag(lambNew) 
    invG = np.eye(d) + Wn.T @ invPsiOld @ Wn 
    Ez = np.linalg.solve(invG, Wn.T @ invPsiOld @ Y) 
    sumEzz = T * np.linalg.inv(invG) + Ez @ Ez.T 

    # M-step
    Wn = (Y @ Ez.T) @ np.linalg.inv(sumEzz) 
    lambNew = 1.0 / np.diag(S - (Wn @ Ez @ Y.T) / T) 

    # Document progress
    covariance_matrix = Wn @ Wn.T + np.diag(1.0 / lambNew + 1e-6)  
    ll[ii] = np.sum([multivariate_normal.logpdf(Y[:, t], mean=np.zeros(n), cov=covariance_matrix) for t in range(T)])

# Verify we are getting the correct result
WW = W @ W.T
WnWn = Wn @ Wn.T
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.loglog([np.min(lamb), np.max(lamb)], [np.min(lamb), np.max(lamb)], 'k', linewidth=2)
plt.scatter(lamb, lambNew, edgecolor='b', facecolor='b', alpha=0.5)
plt.title(r'$\lambda$')
plt.xlabel('true')
plt.ylabel('estimated')
plt.axis('equal')

plt.subplot(1, 2, 2)
plt.plot([np.min(WW), np.max(WW)], [np.min(WW), np.max(WW)], 'k', linewidth=2)
plt.scatter(WW.flatten(), WnWn.flatten(), edgecolor='r', facecolor='r', alpha=0.2)
plt.title(r'$W W^T$')
plt.xlabel('true')
plt.ylabel('estimated')
plt.axis('equal')

# Verify marignal log-likelihood is increasing every step
plt.figure()
plt.plot(range(1, len(ll) + 1), ll, '-o', linewidth=2)
plt.ylabel('marginal log-likelihood')
plt.xlabel('iteration')
```
